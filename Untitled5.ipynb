```python
# Embedding using Word2Vec
import gensim.downloader as api

model = api.load("word2vec-google-news-300")  # Google News에서 사전 훈련된 Word2Vec 모델 로드

color_vectors = ["red", "blue", "yellow", "green", "violet", "cyan", "black", "white"]  # 색상 리스트
color_embeddings = {}  # 임베딩을 저장할 딕셔너리

for vector in color_vectors:  # 각 색상 단어에 대한 임베딩 가져오기
    if vector in model:
        color_embeddings[vector] = model[vector]
    else:
        print(f"Word '{vector}' not in the vocabulary of the model.")

for vector, embedding in color_embeddings.items():  # 임베딩 출력
    print(f"Embedding for '{vector}':\n{embedding}\n")
```

```python
# Embedding using BERT
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# List of color words
color_words = ["red", "blue", "yellow", "green", "violet", "cyan", "black", "white"]

# Dictionary to store embeddings
color_embeddings = {}

# Function to get embeddings for a word
def get_word_embedding(word):
    inputs = tokenizer(word, return_tensors='pt')
    outputs = model(**inputs)
    # Get the embeddings for the [CLS] token (representing the word)
    embedding = outputs.last_hidden_state[:, 0, :].detach().numpy()
    return embedding

# Get embeddings for each color word
for word in color_words:
    color_embeddings[word] = get_word_embedding(word)

# Print the embeddings
for word, embedding in color_embeddings.items():
    print(f"Embedding for '{word}':\n{embedding}\n")
```

```python
# Visualizing embedding results using t-SNE
import torch
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

color_embeddings = np.array([get_word_embedding(word).flatten() for word in color_words])
tsne = TSNE(n_components=2, random_state=42, perplexity=5)  # Set perplexity to 5 (less than 8)
color_embeddings_2d = tsne.fit_transform(color_embeddings)

plt.figure(figsize=(10, 6))  # Plot the 2D t-SNE representation
for i, word in enumerate(color_words):
    plt.scatter(color_embeddings_2d[i, 0], color_embeddings_2d[i, 1])
    plt.text(color_embeddings_2d[i, 0] + 0.02, color_embeddings_2d[i, 1] + 0.02, word, fontsize=12)

plt.title("t-SNE Visualization of Color Word Embeddings")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.show()
```

```python
# Customzied embedding
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np

# List of color words
color_words = ["red", "blue", "yellow", "green", "violet", "cyan", "black", "white"]

# Create a mapping from words to indices
word_to_idx = {word: idx for idx, word in enumerate(color_words)}

# Define a simple embedding model
class ColorEmbeddingModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim):
        super(ColorEmbeddingModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)

    def forward(self, x):
        return self.embedding(x)

# 초기화
# Hyperparameters
vocab_size = len(color_words)
embedding_dim = 2

# Initialize the model, loss function, and optimizer
model = ColorEmbeddingModel(vocab_size, embedding_dim)
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training data: Just indices of words
inputs = torch.tensor([word_to_idx[word] for word in color_words], dtype=torch.long)

# Dummy target to train the model (same as inputs)
targets = inputs.clone().detach()

# Train the model
for epoch in range(1000):
    model.train()
    optimizer.zero_grad()
    outputs = model(inputs)
    loss = criterion(outputs, targets.float().unsqueeze(1))
    loss.backward()
    optimizer.step()

# 결과 출력
model.eval()
with torch.no_grad():
    embeddings = model(inputs).numpy()

for item in embeddings:
    print(item)

# 시각화
# Plot the 2D embeddings
plt.figure(figsize=(10, 6))
for i, word in enumerate(color_words):
    plt.scatter(embeddings[i, 0], embeddings[i, 1])
    plt.text(embeddings[i, 0] + 0.02, embeddings[i, 1] + 0.02, word, fontsize=12)

plt.title("2D Embeddings of Color Words")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.show()
```
